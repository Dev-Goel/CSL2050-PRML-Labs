####**6.)** Evaluation Metrics (Inbulit v/s Scratch)

##### Confusion Matrix


```python
inbuilt_matrix_logistic = cm(y_test,logistic_pred)
inbuilt_matrix_decision = cm(y_test,decision_pred)

print("Confusion Matrix for Logistic Regression-based Predictions =>")
print(inbuilt_matrix_logistic)
print("Confusion Matrix for Decision Tree-based Predictions =>")
print(inbuilt_matrix_decision)


def confusion_matrix(test, pred):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    length = test.size   
    for i in range(length):
        if((test[i] == 0) and (pred[i] == 0)):
            TN = TN + 1
        elif((test[i] == 0) and (pred[i] == 1)):
            FP = FP + 1
        elif((test[i] == 1) and (pred[i] == 0)):
            FN = FN + 1
        else:
            TP = TP + 1
    confusion_matrix = np.array([[TN, FP], [FN, TP]])
    return confusion_matrix

print("\nConfusion Matrix after implementing from scratch\n")
print("Confusion Matrix for Logistic Regression-based Predictions =>")
print(confusion_matrix(y_test, logistic_pred))
print("Confusion Matrix for Decision Tree-based Predictions =>")
print(confusion_matrix(y_test, decision_pred))
```

#####Average Accuracy

```python
inbuilt_acc_logistic = acc(y_test,logistic_pred)
inbuilt_acc_decision = acc(y_test,decision_pred)

print("Accuracy for Logistic Regression-based Predictions =>",str(inbuilt_acc_logistic*100)+"%")
print("Accuracy for Decision Tree-based Predictions =>",str(inbuilt_acc_decision*100)+"%")

def avg_accuracy(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    TN = confusionMatrix[0][0]
    FP = confusionMatrix[0][1]
    FN = confusionMatrix[1][0]
    TP = confusionMatrix[1][1]
    averageAccuracy = (TN + TP)/(TN + TP + FN + FP)
    return averageAccuracy

print("\nAccuracy after implementing from scratch\n")
print("Accuracy for Logistic Regression-based Predictions =>",str(avg_accuracy(y_test, logistic_pred)*100)+"%")
print("Accuracy for Decision Tree-based Predictions =>",str(avg_accuracy(y_test, decision_pred)*100)+"%")
```

#####Precision

```python
inbuilt_ps_logistic = ps(y_test,logistic_pred)
inbuilt_ps_decision = ps(y_test,decision_pred)

print("Precision for Logistic Regression-based Predictions =>",str(inbuilt_ps_logistic*100)+"%")
print("Precision for Decision Tree-based Predictions =>",str(inbuilt_ps_decision*100)+"%")

def precision(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    TP = confusionMatrix[1][1]
    FP = confusionMatrix[0][1]
    precision = TP/(TP + FP)
    return precision

print("\nPrecision after implementing from scratch\n")
print("Precision for Logistic Regression-based Predictions =>",str(precision(y_test, logistic_pred)*100)+"%")
print("Precision for Decision Tree-based Predictions =>",str(precision(y_test, decision_pred)*100)+"%")
```

#####Recall

```python
inbuilt_rs_logistic = rs(y_test,logistic_pred)
inbuilt_rs_decision = rs(y_test,decision_pred)

print("Recall for Logistic Regression-based Predictions =>",str(inbuilt_rs_logistic*100)+"%")
print("Recall for Decision Tree-based Predictions =>",str(inbuilt_rs_decision*100)+"%")

def recall(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    TP = confusionMatrix[1][1]
    FN = confusionMatrix[1][0]
    recall = (TP)/(TP + FN)
    return recall

print("\nRecall after implementing from scratch\n")
print("Recall for Logistic Regression-based Predictions =>",str(recall(y_test, logistic_pred)*100)+"%")
print("Recall for Decision Tree-based Predictions =>",str(recall(y_test, decision_pred)*100)+"%")
```

#####F-1 Score

```python
inbuilt_f1s_logistic = f1s(y_test,logistic_pred)
inbuilt_f1s_decision = f1s(y_test,decision_pred)

print("F1-Score for Logistic Regression-based Predictions =>",str(inbuilt_f1s_logistic*100)+"%")
print("F1-Score for Decision Tree-based Predictions =>",str(inbuilt_f1s_decision*100)+"%")

def f1_score(test, pred):
    _precision = precision(test, pred)
    _recall  = recall(test, pred)
    f1Score = (2*_precision*_recall)/(_precision + _recall)
    return f1Score

print("\nF1-Score after implementing from scratch\n")
print("F1-Score for Logistic Regression-based Predictions =>",str(f1_score(y_test, logistic_pred)*100)+"%")
print("F1-Score for Decision Tree-based Predictions =>",str(f1_score(y_test, decision_pred)*100)+"%")
```

#####Class-Wise Accuracy

```python
# Class-wise classification accuracy
def class_accuracy(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    TP = confusionMatrix[1][1]
    FP = confusionMatrix[0][1]
    FN = confusionMatrix[1][0]
    TN = confusionMatrix[0][0]
    class_wise_accuracy = (TN/(TN+FP) + TP/(TP+FN))/2
    return class_wise_accuracy

print("Class-wise classification accuracy after implementing from scratch\n")
print("Class-wise classification accuracy for Logistic Regression-based Predictions =>",str(class_accuracy(y_test, logistic_pred)*100)+"%")
print("Class-wise classification accuracy for Decision Tree-based Predictions =>",str(class_accuracy(y_test, decision_pred)*100)+"%")
```

#####Sensitivity

```python
# Sensitivity 

def sensitivity(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    TP = confusionMatrix[1][1]
    FN = confusionMatrix[1][0]
    _sensitivity = TP/(TP+FN)
    return _sensitivity

print("Sensitivity after implementing from scratch\n")
print("Sensitivity for Logistic Regression-based Predictions =>",str(sensitivity(y_test, logistic_pred)*100)+"%")
print("Sensitivity for Decision Tree-based Predictions =>",str(sensitivity(y_test, decision_pred)*100)+"%")
```

#####Specificity

```python
def specificity(test, pred):
    confusionMatrix = confusion_matrix(test, pred)
    FP = confusionMatrix[0][1]
    TN = confusionMatrix[0][0]
    _specificity = TN/(FP+TN)
    return _specificity

print("Specificity after implementing from scratch\n")
print("Specificity for Logistic Regression-based Predictions =>",str(specificity(y_test, logistic_pred)*100)+"%")
print("Specificity for Decision Tree-based Predictions =>",str(specificity(y_test, decision_pred)*100)+"%")
```
